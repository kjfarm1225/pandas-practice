```python
import pandas as pd 
```

```python
df = pd.DataFrame(
    [["Alice", 17], ["Bob", 24], ["Carol", 29]],
    index=[10, 20, 30],
    columns=["Name", "Age"],
)
df
```

```python
print(df.loc[10])
print("-----")
print(df.iloc[0])
```

```python
print(df.loc[:, "Name"])
print("-----")
print(df.iloc[:, 0])
```

```python
print(df.loc[[20, 30], ["Name", "Age"]])
print("-----")
print(df.iloc[[1, 2], [0, 1]])
```
