######  pandas groupby + transform 練習

```python
import pandas as pd
```

###### STEP1 データ作成 顧客ごとの購入データを作る

```python
data = {
    "顧客": ["A", "A", "A", "B", "B", "C", "C", "C"],
    "購入金額": [100, 200, 150, 300, 250, 80, 120,160]
}

df = pd.DataFrame(data)
df
```

###### STEP2 groupby: 顧客ごとにグループ化する

```python
grouped = df.groupby("顧客")
```

###### STEP3 累積購入回数:cumcount()を使う 0から始まるので +1 する

```python
df["購入回数"] = grouped.cumcount() + 1
df
```

###### STEP4 平均との差 : 平均との差 = 自分 − 平均

```python
def diff_mean(sr):
    return sr - sr.mean()
df["平均との差"] = grouped["購入金額"].transform(diff_mean)
df
```
